```python
# Labelling the Trained Data
from os import listdir
from os.path import isfile, join
import pickle
labeled_images = []

#labels = ['again','cool','dont know','fingerspell','from','have to','her his','here','learning','leave','live','mine','now','oh i see','our','she he','sign','their','them','we','what','when','where','who','why','your']
labels = ['again','cool','dont know','fingerspell','from','have to','her his','their','why','your']

for l in labels:
    mypath = "/Users/pkaur/tensorflow/SignLanguage/SignDataTrain/" + l;
    file_names = [f for f in listdir(mypath)]
    for fname in file_names:
        list_content = []
        if fname != '.DS_Store':
            list_content.append(fname)
            list_content.append(l)
            labeled_images.append(list_content)

with open('/Users/pkaur/tensorflow/SignLanguage/labeled-images.pkl', 'wb') as fout:
    pickle.dump(labeled_images, fout)
```
